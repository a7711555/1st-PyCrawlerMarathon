2. 取出今天下午三點到五點的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

```python
from datetime import datetime
from datetime import timedelta
import requests

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
def get_all_news(q_date=None):
    url = f'https://www.ettoday.net/news/news-list{( "-" + q_date.strftime("%Y-%m-%d") + "-0" if q_date else "" )}.htm'

    browser = webdriver.Chrome(executable_path='./chromedriver.exe')
    browser.get(url)
    
    last_height = browser.execute_script("return document.body.scrollHeight")

    while True:
        browser.execute_script("window.scrollBy(0, 10000);")
        
        try:
            wait = WebDriverWait(browser, 10)
            wait.until(EC.visibility_of_element_located((By.CLASS_NAME,'loading')))
            wait.until(EC.invisibility_of_element_located((By.CLASS_NAME,'loading')))
            
        except:
            pass

        new_height = browser.execute_script("return document.body.scrollHeight")

        if new_height == last_height:
            break
        else:
            last_height = new_height
    
        
    
    html_source = browser.page_source

    browser.close()
    
    return BeautifulSoup(html_source, 'lxml')
     
```

```python
# 1. 取出今天所有的新聞
soup_today = get_all_news()
for d in soup_today.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)


```

```python
# 2. 取出現在時間兩小時內的新聞

two_hour_ago = datetime.now() - timedelta(hours=2)

for d in soup_today.find(class_="part_list_2").find_all('h3'):
    news_datetime = datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M')
    if news_datetime >= two_hour_ago:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
# 3. 根據範例，取出三天前下午三點到五點的新聞
three_days_ago = datetime.now() - timedelta(days=3)
soup = get_all_news(three_days_ago)

for d in soup.find(class_="part_list_2").find_all('h3'):
    news_datetime = datetime.strptime(d.find(class_='date').text, '%Y/%m/%d %H:%M')

    if (news_datetime >= three_days_ago.replace(hour=15) and 
       news_datetime < three_days_ago.replace(hour=17)):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
