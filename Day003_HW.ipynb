```python
import urllib.request
import zipfile
import os, sys

import xmltodict

res = "http://opendata.cwb.gov.tw/govdownload?dataid=F-D0047-093&authorizationkey=rdec-key-123-45678-011121314"
urllib.request.urlretrieve(res, './Data/example.zip')
f = zipfile.ZipFile('./Data/example.zip')
f.extractall('./Data/day003')

dirs = os.listdir('./Data/day003')

data = dict()
if '64_72hr_CH.xml' in dirs:
    with open('./Data/day003/64_72hr_CH.xml', 'rb') as file:
        data = dict(xmltodict.parse(file.read().decode('utf8')))
```

```python
# 1. 請問高雄市有多少地區有溫度資料
locations = data['cwbopendata']['dataset']['locations']['location']

print(f'高雄市共有 {len(locations)} 個地區有溫度資料')
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
for loc in locations:
    print(f"{loc['locationName']}的第一個時間{loc['weatherElement'][0]['time'][0]['dataTime']}溫度為{loc['weatherElement'][0]['time'][0]['elementValue']['value']}{loc['weatherElement'][0]['time'][0]['elementValue']['measures']}")

```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

for data in locations[0]['weatherElement'][0]['time']:
    print(f"時間點{data['dataTime']}，溫度為{data['elementValue']['value']}{data['elementValue']['measures']}")
```
